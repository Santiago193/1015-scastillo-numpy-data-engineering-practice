##Ejercicio 3: Define una función 'resumen(lista)' que retorne un diccionario con: longitud, suma, promedio, mínimo y máximo de la lista.


```python
def resumen(lista):
    dicc = {
        'longitud': len(lista),
        'suma': sum(lista),
        'promedio': sum(lista)/len(lista),
        'minimo': min(lista),
        'maximo': max(lista),
    }
    return dicc

resultadoej3 = [l for l in range(1, 10)]
resultado = resumen(resultadoej3)
print(resultado)
```

##Ejercicio 1: Crea una matriz 5×4 de enteros aleatorios entre 1 y 100 (seed=7).


```python
import numpy as np
np.random.seed(7)
matriz = np.random.randint(1, 101, size=(5, 4))
print("Matriz Original:")
print(matriz)
print("-" * 30)
```

a) Calcula la media, máximo y mínimo de cada columna.


```python
media_col = matriz.mean(axis=0)
max_col = matriz.max(axis=0)
min_col = matriz.min(axis=0)

print("a) Estadísticas por columna:")
print(f"Medias: {media_col}")
print(f"Máximos: {max_col}")
print(f"Mínimos: {min_col}")
print("-" * 30)
```

b) Calcula la suma acumulada de la primera fila.


```python
suma_acum_fila1 = np.cumsum(matriz[0, :])

print(f"b) Suma acumulada de la primera fila: {suma_acum_fila1}")
print("-" * 30)
```

c) Normaliza la matriz para que cada columna tenga valores entre 0 y 1.
Fórmula: (x - min_col) / (max_col - min_col)



```python
matriz_normalizada = (matriz - min_col) / (max_col - min_col)

print("c) Matriz normalizada (valores entre 0 y 1):")
print(matriz_normalizada)
```

##Ejercicio 4: Crea dos matrices 3×4 y concaténalas vertical y horizontalmente.
##Luego divide el resultado vertical en 3 partes iguales con vsplit.

```python
import numpy as np

A = np.zeros((3, 4)) + 1
B = np.zeros((3, 4)) + 2

vertical = np.vstack((A, B))

horizontal = np.hstack((A, B))

print("Concatenación Vertical (Shape: 6x4):")
print(vertical)
print("\nConcatenación Horizontal (Shape: 3x8):")
print(horizontal)
print("-" * 30)

partes = np.vsplit(vertical, 3)

print("Resultado de vsplit (3 partes de 2x4 cada una):")
for i, p in enumerate(partes):
    print(f"Parte {i+1}:\n{p}")
```
